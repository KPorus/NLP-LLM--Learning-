#Linear regression using Gradient decent with demo datasets

```python
import pandas as pd
import math
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/3_gradient_descent/Exercise/test_scores.csv')
df
```

```python

def gradient_descent(x,y):
  m = 0
  c = 0
  lr = 0.0001 # Reduced learning rate
  n = len(x) # Changed from len(df) to len(x)
  cost_previous = 0
  for i in range(10000): # Increased number of iterations
      yp = m*x +c
      cost = (1/n)*sum([val**2 for val in (y-yp)])
      dm = (-2/n)*sum(x*(y-yp))
      dc = (-2/n)*sum(y-yp)
      if math.isclose(cost, cost_previous, rel_tol=1e-20):
            break
      cost_previous = cost
      m = m - lr*dm
      c = c - lr*dc
      #print("m {}, c {}, cost {} iteration {}".format(m,c,cost, i))

  return m,c

x = df['math']
y = df['cs']
m,b = gradient_descent(x,y)
print(m,b)
```
