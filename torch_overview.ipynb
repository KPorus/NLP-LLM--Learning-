#weight * X = Y

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to plot the linear relationship
def plot_linear_relationship(weight, bias):
    # Generate some data points
    x = np.linspace(-10, 10, 100)
    y = weight * x + bias

    # Create the plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=f'weight={weight}, bias={bias}')
    plt.scatter(x, y, color='red')
    plt.axhline(0, color='gray', lw=0.5)
    plt.axvline(0, color='gray', lw=0.5)
    plt.title('Linear Transformation: y = weight * x + bias')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plotting different weights and biases
plot_linear_relationship(1, 0)  # y = x
plot_linear_relationship(2, 0)  # y = 2x
plot_linear_relationship(0.5, 0)  # y = 0.5x
plot_linear_relationship(1, 5)  # y = x + 5
plot_linear_relationship(1, -5)  # y = x - 5
plot_linear_relationship(-1, 0)  # y = -x
```
