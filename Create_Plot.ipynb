#Plot

```python
import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patch

verts = []
steps = []

points = [(0., 0.), (2., 1.), (3., 3.), (2., 4.), (0., 3.), (0., 0.)]
for i in points:
    verts.append(i)
    if i == points[0]:  # Use MOVETO for the first point
        steps.append(Path.MOVETO)
    elif i == points[-1]:  # Use CLOSEPOLY for the last point
        steps.append(Path.CLOSEPOLY)
    else:  # Use LINETO for all other points
        steps.append(Path.LINETO)

fig, ax = plt.subplots()
plt.axis([-0, 10, -0, 10])
ax.add_patch(patch.PathPatch(Path(verts, steps), facecolor='blue' , lw=2))
plt.show()
```

#cos, sin, tan

```python
a=30;
s = math.sin(math.radians(a));
c = math.cos(math.radians(a));
t = math.tan(math.radians(a));
print("sin value at ",a," : ", round(s,4))
print("cos value at ",a," : ", round(c,4))
print("tan value at ",a," : ", round(t,4))

print("sin degree: ", round(math.degrees(math.asin(4/5)),4));
print("cos degrees: ",round(math.degrees(math.acos(4/5)),4));
print("tan degree: " ,round(math.degrees(math.atan(4/5)),4));

radian = round(math.asin(4/5),4)
print("radian: ", radian)
def toDeg(i):
  return (i*180)/(math.pi);

def toRad(i):
  return (i*math.pi)/180;

print("to deg function: ",round(toDeg(radian),4))
print("to rad: ",round(toRad(toDeg(radian)),4))
```

#sin transformation

```python
from matplotlib import pyplot as plt
import numpy as np
import math

xmin = -2
xmax = 7
ymin = -2
ymax = 2

points = 6*(xmax-xmin);
x = np.linspace(xmin,xmax,points);

def plot_fun():
  fig, ax = plt.subplots();
  plt.axis([xmin,xmax,ymin,ymax]);
  plt.plot([xmin,xmax],[0,0],'b');
  plt.plot([0,0],[ymin,ymax],'b');

plot_fun()
y = np.sin(x)
y1 = np.sin(2*x) # 2pi/2
y2 = 2*np.sin(x) # amplitude
y3 = np.sin(x-(math.pi/2))
plt.plot(x,y,'r')
plt.plot(x,y1,'g')
plt.plot(x,y2,'k')
plt.plot(x,y3,'s')
plt.show()
```

#sin transformation with interaction

```python
from matplotlib import pyplot as plt
import numpy as np
import math
%matplotlib inline
from ipywidgets import interactive

xmin = -2
xmax = 14
ymin = -5
ymax = 5
points = 6*(xmax-xmin);
x = np.linspace(xmin,xmax,points);

def graph(a,b,c):
  fig,ax = plt.subplots();
  eq=f'{a}sin({b}x)+{c}';
  ax.set_title(eq);
  plt.axis([xmin,xmax,ymin,ymax]);
  plt.plot([xmin,xmax],[0,0],'b');
  plt.plot([0,0],[ymin,ymax],'b');
  y = a*np.sin(b*x)+c;
  plt.plot(x,y,'r')
  plt.show()

ig = interactive(graph,a=(-10,10),b=(-10,10),c=(-10,10));
ig
```

#sin interaction v2

```python
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline
from ipywidgets import interactive

xmin = -2
xmax = 14
ymin = -5
ymax = 5
points = 10*(xmax-xmin);
x = np.linspace(xmin,xmax,points);

def graph(a,b,c,d):
  fig,ax = plt.subplots();
  eq=f'{a}sin({b}x-{c})+{d}';
  ax.set_title(eq);
  plt.axis([xmin,xmax,ymin,ymax]);
  plt.plot([xmin,xmax],[0,0],'k');
  plt.plot([0,0],[ymin,ymax],'k');
  y1 = a*np.sin(x)+d;
  y2 = a*np.sin(b*x-c)+d;
  y3 = y1 + y2
  plt.plot(x,y1,'b')
  plt.plot(x,y2,'g')
  plt.plot(x,y3,'r')
  plt.show()

ig = interactive(graph,a=(-10,10),b=(-10,10),c=(-10,10),d=(-10,10));
ig
```

#Tide and Sin Wave

### part one

```python
import requests
import pandas as pd
import arrow
from scipy.optimize import curve_fit
import numpy as np

start = arrow.now().floor('day')
end = arrow.now().shift(days=1).floor('day')

print( start.to('UTC').timestamp())
print(end.to('UTC').timestamp())

response = requests.get(
  'https://api.stormglass.io/v2/tide/extremes/point',
  params={
    'lat': 5.5333312,
    'lng': 95.1999992,
    'start': start.to('UTC').timestamp(),  # Convert to UTC timestamp
    'end': end.to('UTC').timestamp(),  # Convert to UTC timestamp
  },
  headers={
    'Authorization': '3ad5289a-3aa7-11ef-a49b-0242ac130004-3ad52908-3aa7-11ef-a49b-0242ac130004'
  }
)

# Do something with response data.
json_data = response.json()
tide_data = pd.DataFrame(json_data['data'])
print(tide_data.describe())
tide_data.head(1000)
```

###part two

```python
time = pd.to_datetime(tide_data['time'])
height = tide_data['height']

time = (time-time.min()).dt.total_seconds()/(24*60*60)

def sinusoidual_fuc(time,a,b,c,d):
  return a*np.sin(b*time+c)+d

# Provide an initial guess for the parameters
# initial_guess = [2, 0.5, 0, np.mean(height)]

# Increase the maximum number of iterations
params, _ = curve_fit(sinusoidual_fuc,time,height,maxfev=5000);
prediction = sinusoidual_fuc(time,*params)

plt.plot(time,height,label="Actual Tide")
plt.plot(time,prediction,label="Prediction tide")
plt.xlabel("Time")
plt.ylabel("Height(m)")
plt.legend()
plt.show()
```
